```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
import time

url = ""
```

```python
brands = ["Toyota", "Alfa Romeo", "Audi", "BMW", "Buick", "Cadillac", "Chevrolet", "Chrysler", "Citroën", "Cupra", "Dacia",
          "Dodge", "Ferrari", "Fiat", "Ford", "Honda", "Hyundai", "Jaguar", "Jeep", "Kia", "Land Rover", "Lexus", "Lotus",
          "Maserati", "Mazda", "Mercedes", "MG", "Mini", "Mitsubishi", "Morgan", "Morris", "Nissan", "Opel", "Peugeot",
          "Polestar", "Pontiac", "Porsche", "Renault", "Rolls Royce", "Rover", "Saab", "Seat", "Skoda", "Subaru", "Suzuki",
          "Tesla", "Triumph", "Volvo", "VW"] # Alle brands vi vil scrape inde fra autouncle.dk
car_links = [] # Lav en tom liste som vores links skal gemmes i

for brand in brands:
    for i in range(1, 101):
        url = f"https://www.autouncle.dk/da/brugte-biler/{brand}?page={i}&s%5Bmin_price%5D=5000&s%5Bnot_leasing%5D=false&search_form=1" # En dynamisk URL som ændrer sig for at hver side og brand vi går igennem
        page = requests.get(url)
        soup = BeautifulSoup(page.content, 'html.parser')
        car_row = soup.find_all('div', class_='listing-item')
        if soup.find('input', value='no_results_page'): # En stopper, hvilket stopper loopet hvis der ikke er flere biler tilbage. F.eks. hvis der kun er 50 biler vil der kun være 3 sider, men vores range er 1-100. Formålet er at gøre scriptet hurtigere
            break
        for car in car_row: # Opdeler nu hver side i de biler der bliver vist
            car_listed = car.find('a', class_='listing-item-details-link') # Vi fortæller den hvordan id'et vil stå i html'en
            car_id = car_listed.get('href')
            link = f"https://www.autouncle.dk{car_id}" #Vi parser id'en til urlen for at få det fulde link
            car_links.append(link) # Vi gemmer nu linket i den liste vi lavede i starten
        print(brand, i) # Den her funktion har kun til formål at vise hvor langt scriptet er i processen
        time.sleep(2) # For at undgå "too many requests"/ at blive blokeret af hjemmesiden for at spamme den
df = pd.DataFrame(car_links, columns=['Link']) #Vi gemmer vores liste i en dataframe
df.to_csv('car_links.csv', index=False) # Gemmer nu listen som en csv fil. Det tog 48 minutter men i kan evt. selv prøve at køre det hvis i har lyst
```
